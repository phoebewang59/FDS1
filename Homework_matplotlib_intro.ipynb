**3. Basic Line Plot:**


```python
# a) generate a list of numbers from -10 to 10, incremented by 1

new_x = []
for i in range(-11, 10):
    new_x.append(i+1)

new_x = np.array(new_x) #convert list to numpy array
print(new_x)
```

```python
# b) subplot object and plot the numbers against their cube values
new_y = new_x**3

fig , ax = plt.subplots(figsize = (2, 2))

ax.plot(new_x, new_y, label = 'Cubic')
ax.set_xlabel('Y', fontsize = 10)
ax.set_ylabel('Y^3', fontsize = 10)
ax.set_title('Plot of Y vs Y^3', fontsize = 15)
```

**4. Plot Annotations (using ax):**

```python
# b) annotate the point (-5, -125) with the text "This is (-5,-125)"
# c) highlight the point (-5, -125) with a blue circle.

fig , ax = plt.subplots()
ax.plot(new_x, new_y, label = 'Cubic')
ax.set_xlabel('Y', fontsize = 15)
ax.set_ylabel('Y^3', fontsize = 15)
ax.set_title('Plot of Y vs Y^3', fontsize = 20)
ax.text(-5, -125, 'This is (-5, -125)') # annotation
ax.scatter(-5, -125, marker = 'o')      # highlight
```
