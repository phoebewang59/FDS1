**Problem 1: Noisy Phase-Shifted Sine Waves**
---

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Task 1: Generate two noisy sine waves that represent the levels of two different hormones over time.**

```python
#noise level
mean = 0
sd = 0.1

hours = 24
freq = 1/hours                  # how many cycles we want per unit of time (1 cycle in 24 hours)
p = np.random.rand()         # random phase shift
t = np.linspace(0,hours, 1000)  # time, one cycle

h1 = np.sin(2 * np.pi * freq * t) + np.random.normal(mean, sd, 1000)
h2 = np.sin(2 * np.pi * freq * (t + p)) + np.random.normal(mean, sd, 1000)
```

```python
plt.plot(t,h1, label = 'Hormone 1')
plt.plot(t,h2, label = 'Hormone 2')
plt.xlabel('Time (hours)')
plt.ylabel('Hormone Level')
plt.title('Noisy Sine Waves of Hormone Levels Over a Day')
plt.legend()
plt.grid()
```

**Task 2: Plot the two sine waves on the same graph over a period of 10 days.**

```python
#noise level
mean = 0
sd = 0.1

days = 10
hours = 24
freq = 1/hours                      # how many cycles we want per unit of time (1 cycle in 24 hours)
p = np.random.rand()                 # random phase shift
t = np.linspace(0,days*hours, 1000)  # time, one cycle

# noisy sine waves
h1_10days = np.sin(2 * np.pi * freq * t) + np.random.normal(mean, sd, 1000)
h2_10days = np.sin(2 * np.pi * freq * (t + p)) + np.random.normal(mean, sd, 1000)


#plot sine waves
plt.plot(t,h1_10days, label = 'Hormone 1')
plt.plot(t,h2_10days, label = 'Hormone 2')

# labels
plt.xlabel('Time (hours)')
plt.ylabel('Hormone Level')
plt.title('Noisy Sine Waves of Hormone Levels Over Ten Days')
plt.legend()
plt.grid()
```

**Problem 2: Correlated Data and Marginal Distributions**
---

```python
np.random.seed(42) #set seed for reproducibility (generate random numbers)
num_data = 500     
r = 0.8            # correlation coefficient


# normally distributed residuals
resid_mean = 0
resid_sd = 1
residuals = np.random.normal(resid_mean, resid_sd, num_data)

x = np.random.rand(num_data) * 10

y = x * r + residuals

# Set up subplots
gs_kw = dict(width_ratios=[1.4, 1], 
             height_ratios=[1, 2],
             wspace=0.1, hspace=0.1)                   

fig, ax = plt.subplots(2, 2, figsize=(10, 8), gridspec_kw = gs_kw, layout = 'constrained')
fig.suptitle('Scatterplot with Marginal Distributions', fontsize = 25)
label_size = 20

# scatterplot
ax[1,0].scatter(x, y)

# Histogram for marginal distribution of x
ax[0,0].hist(x, bins=15, color='skyblue', edgecolor='black')
ax[0,0].set_xlabel('Variable X', fontsize = label_size)
ax[0,0].set_ylabel('Frequency', fontsize = label_size)

# Histogram for marginal distribution of y
ax[1,1].hist(y, bins=15, orientation = 'horizontal', color='salmon', edgecolor='black')
ax[1,1].invert_xaxis()
ax[1,1].set_xlabel('Frequency', fontsize = label_size)
ax[1,1].set_ylabel('Variable Y', fontsize = label_size)

# Remove unnecessary plot
fig.delaxes(ax[0,1])
```
